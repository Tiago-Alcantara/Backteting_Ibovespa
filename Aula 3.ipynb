```python
import yfinance as yf
import pandas as pd
import quantstats as qs
```

## **Passo 1** - Ler a composição hsitorica do Ibovespa e os tickers que ja passaram pelo índice.

```python
composicao_historica = pd.read_excel(r'C:\Users\Tiago\Documents\Codigos\bot.camp\Aula_3\composicao_ibov_.xlsx')
lista_acoes = pd.read_excel(r'C:\Users\Tiago\Documents\Codigos\bot.camp\Aula_3\composicao_ibov_.xlsx',sheet_name= 'lista_acoes')

print(lista_acoes)
print(composicao_historica)
```

## **Passo 2** - Puxar as cotações de todas as empresas que farão parte do backtest.

```python
dados_cotacoes = (yf.download(lista_acoes['tickers'].to_list(),
                             start= "2015-06-30", end = "2022-12-31")['Adj Close'])


dados_cotacoes
```

## **Passo 3** - Transformar o índice em data e ordenar a serie de tempo.

```python
dados_cotacoes.index= pd.to_datetime(dados_cotacoes.index)

dados_cotacoes
```

```python
dados_cotacoes = dados_cotacoes.sort_index()

dados_cotacoes
```

## **Passo 4** - Calcular o retorno 6 meses de ajustar a tabela com o fator.

```python
r6 = (dados_cotacoes.resample("M").last().pct_change(periods = 6).dropna(axis = 0, how = "all").drop("2022-12-31"))# periods define a quantidade de meses para fazer a operação 

r6
```

## **Passo 5** - Classificar e retirar empreas que não participam do Ibovespa no período de tempo selecionado. 

```python
for data in r6.index:
    for empresa in r6.columns:

        if empresa.replace(".SA", "") not in composicao_historica.loc[:,data].to_list():

            r6.loc[data, empresa] = pd.NA



r6
```

```python
composicao_historica
```

## **Passo 6** - Criar as carteiras de investimento em uma matriz de 0 ou 1.

```python
rankings = r6.rank(axis= 1,ascending= False).applymap(lambda x: 1 if x < 11 else 0)

rankings
```

## **Passo 7** - Calcular o retorno mensal das empreas no período de backtest

```python
retorno_mensal = dados_cotacoes.resample("M").last().pct_change(periods = 1)

retorno_mensal = retorno_mensal.drop(retorno_mensal.index[:7], axis = 0)

retorno_mensal.index = rankings.index

retorno_mensal
```

## **Passo 8** - Cruzar a matriz de retorno mensal com a matriz das carteiras para chegar na rentabilidade do modelo

```python
retorno_modelo = (retorno_mensal * rankings).sum(axis = 1)/10

retorno_modelo
```

## **Passo 8.2** - Visualizar a rentabilidade do modelo com o quantstats

```python
qs.extend_pandas()

retorno_modelo.plot_monthly_heatmap()
```

## **Passo 9** - Puxar e calcular a rentabilidade do Ibovespa no período.

```python
Ibovespa = yf.download("^BVSP", start= "2015-12-30",end="2022-12-31")["Adj Close"]

Ibovespa = Ibovespa.resample("M").last().pct_change().dropna()

Ibovespa
```

## **Passo 10** - Calcular e visualizar as rentabilidades do modelo contra o Ibovespa

```python
retorno_acumulado = (1+ retorno_modelo).cumprod()-1

retorno_acumulado_ibov =(1+ Ibovespa).cumprod()-1 

retorno_acumulado.plot_monthly_heatmap()

retorno_acumulado_ibov.plot_monthly_heatmap()
```

## **Passo 11** - Calcular e visializar as rentabilidades do modelo mes a mes contra o ibovespa

```python
ganhos_modelo = retorno_modelo - Ibovespa

ganhos_modelo.plot_monthly_heatmap()
```
